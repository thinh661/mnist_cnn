```python
import tensorflow as tf
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

print(tf.__version__)
```

1. Load data

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

2. Get Infor Data

```python
import matplotlib.pyplot as plt

label_counts = np.bincount(y_train)
colors = plt.cm.viridis(np.linspace(0, 1, len(label_counts)))
labels = np.arange(len(label_counts))

plt.figure(figsize=(10, 6))
plt.bar(labels, label_counts, color=colors)
plt.xlabel('Nhãn')
plt.ylabel('Số lượng mẫu')
plt.title('Số lượng mẫu cho mỗi nhãn trong tập dữ liệu huấn luyện')

plt.xticks(labels)

plt.show()
```

4. Normalization Data

```python
input_shape = (28, 28, 1)

x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], x_train.shape[2], 1)
x_train = x_train / 255.0
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], x_test.shape[2], 1)
x_test = x_test / 255.0

# print(x_train[100])
```

```python
y_train = tf.one_hot(y_train.astype(np.int32), depth=10)
y_test = tf.one_hot(y_test.astype(np.int32), depth=10)

plt.imshow(x_train[1000][:, :, 0])
print(y_train[1000])
```

```python
model = tf.keras.models.load_model("model_v1.h5")
score = model.evaluate(x_test, y_test, batch_size=16)
```
